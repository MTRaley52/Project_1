```python
import matplotlib.pyplot as plt
import pandas as pd

#Take in Data
total_energy_prd_cons_df = pd.read_csv('Total_Energy_ProdConsume_QudBTU_FC_Prj1.csv')

#Group and sort the values
total_energy_prd_cons_grouped = total_energy_prd_cons_df.groupby(['ENERGY_TYPE']).sum()
total_energy_prd_cons_grouped = total_energy_prd_cons_grouped.sort_values(
    by=['ProdQuadBTU'], ascending= False)

#Drop year to only focus on Consumption and Production values
total_energy_prd_cons_bar = total_energy_prd_cons_grouped.drop(columns=['Year'])

#Plot the Data
total_energy_prd_cons_bar.plot.bar()
plt.xlabel("Energy Type fc to 2050 ")
plt.ylabel("Quadrillion btu")
plt.title("US Energy Forecast 2018-2050 Production - Consumption  EIA")
plt.grid(True)
plt.show()

#plt.savefig("US Total by Type Energy FC.png")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

#Take in data
ren_elec_fc = pd.read_csv('RenewableElectricFC.csv')

#Sort data and drop Total
ren_elec_fc = ren_elec_fc.sort_values(by=['Year'])
#ren_elec_fc = ren_elec_fc.drop(columns=['Total GW'])

#Set Y-axis as Year 
ren_elec_fc = ren_elec_fc.set_index('Year')
#Plot the data
ax = plt.subplot()
ren_elec_fc.plot(kind = 'line', ax =ax)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title("US Renewable Energy Forecast by Type")
plt.xlabel("Forcast Year")
plt.ylabel("GW hrs ")
plt.grid(True)
plt.show()


#plt.savefig("Renewable Energy FC.png")
```
