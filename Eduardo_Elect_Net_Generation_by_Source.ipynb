```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from functools import reduce
import requests
import json
from pprint import pprint
from confy import gkey
```

```python
#No renewable net generation Coal
base_url = "http://api.eia.gov/series/?api_key=gkey&series_id=ELEC.GEN.COW-NM-98.A"
    #api_key=YOUR_API_KEY_HERE&series_id=SEDS.CLPRB.NM.A
parame={"api_key":gkey,"series_id":"ELEC.GEN.COW-NM-98.A"}
response_json = requests.get(base_url, params=parame).json()

response_json
pd.DataFrame(response_json["series"][0]["data"])
```

```python
NM_net_coal_production=pd.DataFrame(response_json["series"][0]["data"])
print(NM_net_coal_production)
```

```python
NM_net_coal_production.columns=["Year","Coal_Net_Generation"]
print(NM_net_coal_production)
NM_net_coal_production_sort=NM_net_coal_production.sort_values('Year',ascending=True)
NM_net_coal_production_sort
```

```python
xb=NM_net_coal_production_sort["Year"]
yb=NM_net_coal_production_sort["Coal_Net_Generation"]
```

```python
plt.plot(xb,(yb/3.412142), color = "teal", linewidth=3)

axis_font = {'fontname':'Arial', 'size':'18'}
axis_font2 = {'fontname':'Arial', 'size':'16'}

plt.title('Electricity_Net_Generation_Coal_(2001-2018)',**axis_font)
plt.xlabel("Year", **axis_font2)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.2)
plt.ylabel('Billion-Btu', **axis_font2)
#plt.title('Energy (Electricity) Consumption by Source')
plt.grid(True)

plt.rcParams["figure.figsize"] = (14,6)
#plt.ylim(10000,32500)
```
